```python
import librosa
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
import os
```

```python
def extract_features(audio_path):
    # Load audio file with resampling to 44100 Hz
    audio, sr = librosa.load(audio_path, sr=44100)

    # Feature 1: MFCC
    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc, axis=1)

    # Feature 2: Chroma STFT
    chroma_stft = librosa.feature.chroma_stft(y=audio, sr=sr)
    chroma_stft_mean = np.mean(chroma_stft, axis=1)

    # Feature 3: Chroma CQT with reduced bins
    chroma_cqt = librosa.feature.chroma_cqt(y=audio, sr=sr, bins_per_octave=24, n_octaves=7)
    chroma_cqt_mean = np.mean(chroma_cqt, axis=1)

    # Feature 4: Chroma CENS
    chroma_cens = librosa.feature.chroma_cens(y=audio, sr=sr)
    chroma_cens_mean = np.mean(chroma_cens, axis=1)

    # Combine all features into a single vector
    return np.hstack([mfcc_mean, chroma_stft_mean, chroma_cqt_mean, chroma_cens_mean])
```

```python
def process_audio_files(audio_files_directory, output_csv):
    features_list = []
    labels_list = []

    # Traverse through each folder (each object class)
    for folder_name in os.listdir(audio_files_directory):
        folder_path = os.path.join(audio_files_directory, folder_name)
        
        # Ensure it's a directory
        if os.path.isdir(folder_path):
            for audio_file in os.listdir(folder_path):
                if audio_file.endswith(".wav"):  # Ensure it's an audio file
                    audio_file_path = os.path.join(folder_path, audio_file)
                    print(f"Processing {audio_file_path}...")

                    # Extract features and append to the list
                    features = extract_features(audio_file_path)
                    features_list.append(features)
                    
                    # Append the label (which is the folder name)
                    labels_list.append(folder_name)

    # Convert to a DataFrame
    feature_columns = [f"feature_{i}" for i in range(len(features_list[0]))]
    df = pd.DataFrame(features_list, columns=feature_columns)
    
    # Add the labels column
    df['label'] = labels_list

    # Save to CSV
    df.to_csv(output_csv, index=False)
    print(f"Features and labels saved to {output_csv}")
```

```python
# Define the path to your audio files directory
audio_files_directory = "/home/aditya/ADITYA/Projects/Denoise/sound_datasets/urbansound8k/audio"
output_csv = 'audio_features_with_labels.csv'

# Process the audio files and save features with labels to the CSV
process_audio_files(audio_files_directory, output_csv)
```
